# Analyzing dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
import ast
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
import re

def extract_sota_papers(text):
    pattern = r'"(.*?)" \(\d{4}\)'
    return re.findall(pattern, str(text))
```

```python
df['num_sota_papers'] = df['SoTA Results'].apply(lambda x: len(extract_sota_papers(x)))
```

```python
def count_critique_categories(critique_dict):
    return len(critique_dict.keys())
```

```python
df['Combined Critiques'] = df['Combined Critiques'].apply(ast.literal_eval)
df['num_critique_categories'] = df['Combined Critiques'].apply(count_critique_categories)
```

```python
def count_evidence_references(evidence_dict):
    return sum(len(v.split('\n')) for v in evidence_dict.values())
```

```python
df['Retrieved Evidence'] = df['Retrieved Evidence'].apply(ast.literal_eval)
df['num_evidence_references'] = df['Retrieved Evidence'].apply(count_evidence_references)
```

# Visualizations

```python
# Relationship 1: SoTA Papers vs. Critique Categories
plt.figure(figsize=(8, 6))
sns.regplot(x=df['num_sota_papers'], y=df['num_critique_categories'])
plt.xlabel("Number of SoTA Papers")
plt.ylabel("Number of Critique Categories")
plt.title("Relationship between SoTA Papers and Critique Categories")
plt.show()
```

```python
# Relationship 2: Critique Categories vs. Evidence References
plt.figure(figsize=(8, 6))
sns.regplot(x=df['num_critique_categories'], y=df['num_evidence_references'])
plt.xlabel("Number of Critique Categories")
plt.ylabel("Number of Evidence References")
plt.title("Relationship between Critique Categories and Evidence References")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud

# Word Cloud for Critique Topics
all_critiques = " ".join(df['Combined Critiques'].dropna().astype(str))
vectorizer = CountVectorizer(stop_words='english')
word_freq = vectorizer.fit_transform([all_critiques])
word_counts = dict(zip(vectorizer.get_feature_names_out(), word_freq.toarray()[0]))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Common Words in Critiques")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['num_sota_papers', 'num_critique_categories', 'num_evidence_references']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
